```python
# The follwoing section to remove null values
#create a new Dataframe which would hold clean data
player_stats_201617_clean_df = player_stats_201617_df
```

```python
# Get the count of rows
player_stats_201617_clean_df.count()
```

```python
# get the count of null values in each column
player_stats_201617_clean_df.isnull().sum()
```

```python
#List the columns where NAN values needs to be replace with 0
player_stats_201617_clean_df[['FG%','3P%','2P%','eFG%','FT%']]
```

```python
#summarize NAN count
player_stats_201617_clean_df[['FG%','3P%','2P%','eFG%','FT%']].isnull().sum()
```

```python
#replace nulll values with 0
player_stats_201617_clean_df[['FG%','3P%','2P%','eFG%','FT%']] = player_stats_201617_clean_df[['FG%','3P%','2P%','eFG%','FT%']].fillna(value=0)
```

```python
# check whether the null values are replaced in new dataframe
player_stats_201617_clean_df[['FG%','3P%','2P%','eFG%','FT%']].isnull().sum()
```

```python
# Put player name into a list
player_id_name = player_stats_201617_clean_df['Player'].tolist()
```

```python
# Split the playid and name and list the rows which are not split into 2 parts  with \ as delimeter(either 1 or >2)
for name in player_id_name:
    print(name.split("\\"), len(name.split("\\")))
```

```python
# Split the playid and name and list the rows which are not split into 2 parts (either 1 or >2)
for name in player_id_name:
    if (len(name.split("\\")) != 2):
        print(name.split("\\"), len(name.split("\\")))
```

```python
# adjust split argument to check whether our code wortks by changing delimeter to space
# Split the playid and name and list the rows which are not split into 2 parts (either 1 or >2)
for name in player_id_name:
    if len(name.split()) != 2:
        print(name.split(), len(name.split()))
# could see that all rows are split into 2 (not less, not more)
```

```python
# Splits the Name Column into two columns and add it to Dataframe
split_name_df = player_stats_201617_clean_df['Player'].str.split("\\", n = 1, expand = True)
split_name_df
```

```python
split_name_df.dtypes
```

```python
#Add New column called player_name
player_stats_201617_clean_df["player_name"] = split_name_df[0]
player_stats_201617_clean_df["player_id"] = split_name_df[1]
```
